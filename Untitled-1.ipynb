```python
wd=pd.read_csv(r"D:\archive\worldometer_data.csv")
wd.head()
```

```python
wd.isnull().sum()
```

```python
wd.fillna({"NewCases" : wd["NewCases"].mean()},inplace=True)
wd.fillna({"TotalDeaths" : wd["TotalDeaths"].mean()},inplace=True)
wd.fillna({"NewDeaths" : wd["NewDeaths"].mean()},inplace=True)
wd.fillna({"TotalRecovered" : wd["TotalRecovered"].mean()},inplace=True)
wd.fillna({"NewRecovered" : wd["NewRecovered"].mean()},inplace=True)
wd.fillna({"ActiveCases" : wd["ActiveCases"].mean()},inplace=True)
wd.fillna({"Serious,Critical" : wd["Serious,Critical"].mean()},inplace=True)
wd.fillna({"Tot Cases/1M pop" : wd["Tot Cases/1M pop"].mean()},inplace=True)
wd.fillna({"Deaths/1M pop" : wd["Deaths/1M pop"].mean()},inplace=True)
wd.fillna({"TotalTests" : wd["TotalTests"].mean()},inplace=True)
wd.fillna({"Tests/1M pop" : wd["Tests/1M pop"].mean()},inplace=True)
wd.head()
```

```python
#wd.isnull().sum()
#wd["WHO Region"].value_counts()
wd.fillna({"WHO Region" : "Europe"},inplace=True)
wd.head()
```

```python
plt.bar(wd["WHO Region"],wd["TotalDeaths"])
plt.xlabel("region")
plt.ylabel("Victims")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.bar(wd["WHO Region"],wd["TotalRecovered"])
plt.xlabel("region")
plt.ylabel("recovered")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.plot(wd["TotalCases"],wd["TotalRecovered"])
plt.xlabel("cases")
plt.ylabel("recovered")
plt.show()
```

```python
plt.plot(wd["TotalCases"],wd["TotalDeaths"])
plt.xlabel("cases")
plt.ylabel("infected")
plt.show()
```
